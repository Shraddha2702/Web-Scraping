```python
def deal_with_string(word):
    word = word.lower()
    for punn in punc: 
        word = word.replace(punn, '')
        word = word.strip()
    return word
```

```python
deal_with_string('y\'s clich&eacute;s ')
```

```python
def convert_list_to_xml(data):
    whole = ""
    for i in range(len(data)):
        if(type(data[i]) == str):
            whole += "<item>"+deal_with_string(data[i])+"</item>"
        elif(type(data[i]) == int or type(data[i]) == None):
            whole += "<item>"+str(data[i])+"</item>"
        elif(type(data[i]) == list):
            whole += "<item>"
            whole += convert_list_to_xml(data[i])
            whole += "</item>"
        else:
            whole += "<item>"
            whole += convert_dic_to_xml(data[i])
            whole += "</item>"
    return whole
```

```python
convert_list_to_xml(['a','b','c',['1', '2', '3']])
```

```python
def convert_dic_to_xml(data):
    whole = ""
    for each in data.keys():
        whole += "<"+each+">"
        if(type(data[each]) == str):
            whole += deal_with_string(data[each])
        elif(type(data[each]) == int or type(data[each]) == None):
            whole += str(data[each])
        elif(type(data[each]) == dict):
            whole += convert_dic_to_xml(data[each])
        elif(type(data[each]) == list):
            whole += convert_list_to_xml(data[each])
        whole += "</"+each+">"
    return whole
```

```python
with open('stopwords.txt', 'r') as f:
    content = f.readlines()

# you may also want to remove whitespace characters like `\n` at the end of each line
final_list = [x.strip() for x in content]
punc = [",", ";", ".", "/","\"", "$", "%", "^", "&", "*", "(", ")", "-","_", "+", "=", "{", "}","[", "]", "*", ";", "?","<i>","</i>","#"]
final_list.append(punc)
```

```python
#Reading from xml and creating inverted Index
tree = etree.parse("prize1.xml")
```

```python
root = etree.Element("index")

for i in range(len(tree.xpath("//motivation"))):
    em1 = tree.xpath("//id")[i]
    em2 = tree.xpath("//motivation")[i]
    sentence = em2.text
    words = sentence.split()
    ids = em1.text
    
    for word in words:
        worddicc = [x.text for x in root.xpath("//keyword")]
        word = word.lower()
        #print(word, worddicc)
        for punn in punc: word = word.replace(punn, "")
        if((word not in final_list) and (word not in worddicc) and (word != "")):
            #Create a new Tag
            entry = etree.SubElement(root, "entry")
            keyword = etree.SubElement(entry, "keyword")
            keyword.text = word
            idss = etree.SubElement(entry, "ids")
            idsss = etree.SubElement(idss, "id")
            idsss.text = ids
        elif(word in worddicc):
            idss = root.xpath('.//keyword[text()="'+word+'"]')[0].getparent()[1]
            idsss = etree.SubElement(idss, "id")
            idsss.text = ids
```
