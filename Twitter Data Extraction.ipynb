```python
df = pd.read_csv('newtweets.csv')
new_df = pd.DataFrame({'Tweet Created At':df['created_at'],
                      'Tweet Favorited':df['favorite_count'],
                      'Reply or Not':df['in_reply_to_screen_name'].isnull(),
                      'Retweet Count':df['retweet_count'],
                      'Tweet':df['text']})
```

```python
new_um = []
new_media = []
new_ht = []

for i in range(len(df)):
    js = ast.literal_eval(df['entities'][i])
    if('user_mentions' in js.keys()):
        valsm = []
        for elem in js['user_mentions']:
            valsm.append(elem['screen_name'])
        new_um.append(valsm)
    else:
        new_um.append('')
    
    if('media' in js.keys()):
        new_media.append('Yes') #Yes or No
    else:
        new_media.append('No')
    
    if('hashtags' in js.keys()):
        vals = []
        for elem in js['hashtags']:
             vals.append(elem['text'])
        new_ht.append(vals)
    else:
        new_ht.append([])
```

```python
new_ht
```

```python
len(new_um), len(new_media), len(new_ht)
```

```python
new_df.insert(loc=5, column='User mentioned in Tweet', value=new_um)
new_df.insert(loc=6, column='Media File or Not', value=new_media)
new_df.insert(loc=7, column='Hashtags Used', value=new_ht)
```

```python
sn = []
name = []
sc = []
fc = []
favc = []
follc = []
ppurl = []

for i in range(len(df)):
    jsn = ast.literal_eval(df['user'][i])
    sn.append(jsn['screen_name'])
    name.append(jsn['name'])
    sc.append(jsn['statuses_count'])
    fc.append(jsn['friends_count'])
    favc.append(jsn['favourites_count'])
    follc.append(jsn['followers_count'])
    ppurl.append(jsn['profile_image_url_https'])
    
new_df.insert(loc=8, column='Company Name', value=name)
new_df.insert(loc=9, column='Screen Company Name', value=sn)
new_df.insert(loc=10, column='Acc Status Count', value=sc)
new_df.insert(loc=11, column='Acc Friends Count', value=fc)
new_df.insert(loc=12, column='Acc Favorites Count', value=favc)
new_df.insert(loc=13, column='Acc Followers Count', value=follc)
new_df.insert(loc=14, column='Profile Picture URL', value=ppurl)
```

```python
com_df = pd.DataFrame({'SN':sn, 'Name': name, 'SC':sc, 'FC':fc, 'FavC':favc, 'FollowC':follc, 'PPUrl':ppurl})
com_df.to_csv('CompanyInfo.csv', index=False)
```
